## Loading the dataset
read_CSV assumes the data contains a header. In that case we can just pass the path of the dataset.
If data has no column headers. we need to specify read_CSV to not assign headers by setting header to none.

```python
data = pd.read_csv('car_prices.csv', header=None)
```
