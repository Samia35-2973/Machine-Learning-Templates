# Import necessary packages

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

# Import datasets

```python
train_df = pd.read_csv("/content/drive/MyDrive/Datathon Demo/train_data.csv")
test_df = pd.read_csv("/content/drive/MyDrive/Datathon Demo/test_data.csv")
```

```python
train_df.head()
```

```python
train_df.describe()
```

# Preprocessing and Model apply

```python
encode = LabelEncoder()
train_df.Survived = encode.fit_transform(train_df.Survived)
```

...

...

```python
model = LogisticRegression(max_iter = 10000)
train_X = train_df.drop(columns =['Survived'],axis = 1)
train_y = train_df['Survived']
model.fit(train_X,train_y)
```

   .

.



---


# Save and View the Predictions






```python
predictions = model.predict(test_df)
print("PREDICTIONS",encode.inverse_transform(predictions))
```

# Add the unique_id column and the prediction column with corresponding column headers/name to the submission DataFrame. Check the training dataset for the columns names/headers

```python
submission = pd.DataFrame({'PassengerId':test_df['PassengerId'],'Survived':predictions})
submission.head()
```

# Save the submission DataFrame to the submission.csv file

```python
filename = 'submission.csv'
submission.to_csv(filename, index = False)
print("File saved as "+filename)
```
